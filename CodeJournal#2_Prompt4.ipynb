Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

I used the Beta distribution with alpha = 2 and and the beta = 5 

```python
data = np.random.beta(a=2.0,b=5.0,size=1000)
```

Creating the histogram

```python
plt.hist(data, bins=100, edgecolor='Black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers from Beta(2,5) Distribution')
```

```python
plt.savefig('Beta Histogram.pdf')
plt.show
```
