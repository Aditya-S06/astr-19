Importing python libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining function that:
1. Creating numpy x
2. returns exp(x)

```python
def returnExp():
    x = np.linspace(0,1,100)
    return np.exp(x)
```

Setting 'y' to equal exp(x) by using the function 

```python
y = returnExp()
```

Funtion for the plot x vs. y

```python
def plot(x,y):
    flag_save = True
    xlabel = 'Milliseconds'
    ylabel = 'Awesomeness'
    lcolor = 'red'
    fs = 14
    fname = 'Jour. #2, Prompt 1 plot.pdf'
    f, ax = plt.subplots(1,1,figsize = (4,4))
    ax.plot(x,y,color=lcolor,linewidth=1.5)
    ax.set_xlabel(xlabel,fontsize=fs)
    ax.set_ylabel(ylabel,fontsize=fs)
    if (flag_save):
        plt.savefig(fname,bbox_inches='tight',dpi=400)
```

Defining the main function to run the plot

```python
def main():
    print("Plot for Prompt#1")
    x = np.linspace(0,1,100)
    plot(x,y)
```

Running the main function 

```python
if __name__=='__main__':
    main()
```
