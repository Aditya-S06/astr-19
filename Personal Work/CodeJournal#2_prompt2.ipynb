Importing python libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Initializing x

```python
x = np.linspace(0,1,100)
```

Function returning sin(x) and cos(x)

```python
def sinReturn():
    return np.sin(x)

def cosReturn():
    return np.cos(x)
```

Function for the Math Plot 

```python
def multiPanel(x):
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    # Plot sin(x) on the left
    ax[0].plot(x, sinReturn(), color='red', linewidth=1.5, label='sin(x)')
    ax[0].set_title('sin(x)')
    ax[0].set_xlabel('x')
    ax[0].set_ylabel('sin(x)')

    # Plot cos(x) on the right
    ax[1].plot(x, cosReturn(), color='blue', linewidth=1.5, label='cos(x)')
    ax[1].set_title('cos(x)')
    ax[1].set_xlabel('x')
    ax[1].set_ylabel('cos(x)')

    plt.tight_layout()
    plt.savefig("Jour2_Prompt2.pdf", bbox_inches='tight', dpi=500)
    plt.show()
```

Main function

```python
def main():
    print("Sin(x) and Cos(x) plots")
    x = np.linspace(0,1,100)
    multiPanel(x)
```

Telling python to run the main function

```python
if __name__==("__main__"):
    main()
```
