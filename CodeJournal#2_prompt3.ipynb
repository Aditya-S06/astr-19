Import libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generating a 1000 uniform random numbers in [0,1]

```python
x = np.random.uniform(low=0.0, high=1.0, size=1000)
```

Creating the histogram 

```python
plt.hist(x, bins=100,edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Uniform [0,1] Random Numbers')
```

Code to save Histogram as pdf

```python
plt.savefig('Uniform_histogram.pdf')
plt.show()
```
